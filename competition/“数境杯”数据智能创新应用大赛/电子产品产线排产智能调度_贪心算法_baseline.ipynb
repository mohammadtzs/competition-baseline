## 题目描述

在制造业中，生产调度是一个关键的环节，它对生产效率、产品质量和企业竞争力有着重要的影响。排产调度是指在有限的时间和资源的情况下，合理分配产品制造资源并确定工单排产加工的顺序，以完成特定的排产任务并实现预先设定的优化目标。作为企业生产运营的核心物质基础，机器设备若在运行过程中缺乏合理维护，极易引发故障，进而导致多重负面影响：一方面会增加设备维修成本，另一方面可能造成产线中断、在制工件批量报废等问题，最终给企业带来重大经济损失。因此，科学实施设备维护以降低故障发生率，既是企业控制损失的关键举措，也是提升生产效率的核心手段。值得注意的是，在实际生产场景中，设备维护计划往往需要与生产任务调度相结合。例如，如何在保障订单交付周期的前提下优化维护时机，或如何平衡预防性维护成本与故障停机损失等。这种将生产排序调度与预防性维护统筹规划的复杂性，使得相关研究兼具理论深度与实践意义，也因此持续受到学术界与工业界的共同关注。

## 比赛任务

根据产品物料的订单需求数据，智能算法排产一个车间内所有工单在每个设备的加工顺序，排产结果包括每个工单在哪台设备生产和开始生产时间和结束生产时间，和每台设备每天的开始保养时间和结束保养时间。工厂生产指导员根据智能排产结果进行备料与生产。

## 数据描述

根据订单需求数据、工艺路线、工作日历、切换时间等数据，排产某个车间所有工单在每台设备的加工顺序，初赛排产开始时间为（2025-05-06 08:00:00），决赛排产开始时间为（2025-06-08 08:00:00）。排产需满足以下约束：
1. 每台设备只能加工某件产品的一种工序，且一旦开始不能中断。
2. 产品加工顺序必须按照工艺路线执行，一道工序结束后，下一道工序才可以开始。
3. 每一道工序可在可选设备集合中的选择一台设备进行加工，每道工序只能同时在一台设备上加工。
4. 每个工序必须在一台设备的工作日历的一个班次内加工完成，不能跨班次生产，例如：白班2025-04-22 08:00:00到2025-04-22 12:00:00为一个班次。
5. 如果该机器该天有安排工单排产，则该机器需要在该天上午进行保养一次，每次保养需要30分钟，保养截止时间不超过中午12点，如果该机器该天没有安排工单排产，则不需要保养。

## 初赛评分规则

考虑n个工单在m个设备上的生产过程，每个工单i只有一道工序。其中，
- 交付率评分A1=按时交付订单数/总订单数×100
- 均衡生产评分A2=（1-各设备完工时间的标准差/所有设备生产总时长）×100
- 转产时间评分A3=（1-总切换时间/所有设备生产总时长）×100

注：
- 按时交付：订单内所有工单的最大完工时间所在日期≤订单需求日期；
- 所有设备生产总时长不包括因生产日历造成的班次间隔；

排产效果得分A=0.5×交付率评分A1+0.3×均衡生产评分A2+0.2×转产时间评分A3，初赛以A进行排名。

## 决赛评分规则

算法运行时间评分B=（1-算法运行时间/最长算法运行时间）×100

注：(1)最长算法运行时间为一个常量，设置为1个小时

综合评分C=0.6×排产效果得分A+0.4×算法运行时间评分B

决赛以C进行排名。

```python
import pandas as pd
from datetime import datetime, timedelta

class ProductionScheduler:
    def __init__(self, data):
        self.data = data
        self.prepare_data()

    def prepare_data(self):
        """
        预处理数据，转换时间格式，构建方便查询的数据结构。
        """
        # 转换时间字符串为 datetime 对象
        for df_name in ["工作日历", "产品需求", "瓶颈物料"]:
            df = self.data[df_name]
            for col in df.columns:
                if "时间" in col or "日期" in col:
                    # 尝试多种格式以提高兼容性
                    df[col] = pd.to_datetime(df[col], errors='coerce')
        
        # 将工艺路线数据处理成更易于访问的字典
        # key: (工艺路线类型, 工序号), value: list of [设备号, 单位加工时间]
        self.process_routes = {}
        for _, row in self.data["工艺路线"].iterrows():
            key = (row["工艺路线类型"], row["工序号"])
            if key not in self.process_routes:
                self.process_routes[key] = []
            self.process_routes[key].append({
                "device": row["设备号"],
                "time_per_unit": row["每个产品单位加工时间（秒）"]
            })

        # 切换时间字典
        # key: (设备号, 前产品, 后产品)
        self.switch_times = {}
        for _, row in self.data["切换时间"].iterrows():
            key = (row["设备号"], str(row["前产品"]), str(row["后产品"]))
            self.switch_times[key] = timedelta(minutes=row["切换时间（分钟）"])

        # 工作日历字典
        # key: 设备号, value: sorted list of (开始时间, 结束时间)
        self.calendar = {}
        for _, row in self.data["工作日历"].iterrows():
            device = row["设备号"]
            if device not in self.calendar:
                self.calendar[device] = []
            self.calendar[device].append((row["开始时间"], row["结束时间"]))
        for device in self.calendar:
            self.calendar[device].sort()
            
        # 瓶颈物料供应字典
        # key: (物料号, 日期)
        self.material_supply = {}
        for _, row in self.data["瓶颈物料"].iterrows():
            key = (str(int(row["物料号"])), row["供应日期"].date())
            self.material_supply[key] = row["供应数量"]


    def schedule(self, start_time_str="2025-05-06 08:00:00"):
        """
        执行贪心调度算法
        """
        start_time = datetime.strptime(start_time_str, "%Y-%m-%d %H:%M:%S")

        # 初始化设备状态
        devices_status = {
            device: {"available_time": start_time, "last_product": None}
            for device in self.data["工作日历"]["设备号"].unique()
        }
        
        # 初始化物料使用情况
        material_usage = {} # key: (物料号, 日期), value: 使用量
        
        # 记录已完成的工单的前置任务完成时间
        completed_jobs_time = {} # key: 工单号, value: 结束时间

        # 待排产工单池 (只放入第一道工序)
        pending_jobs = self.data["产品需求"][self.data["产品需求"]["工序号"] == 1].copy()
        
        # 最终排产结果
        schedule_result = []
        
        # 已安排保养的设备和日期
        maintenance_scheduled = set() # key: (设备号, 日期)

        # 主循环
        while not pending_jobs.empty:
            # 贪心策略：优先处理需求日期最早的工单
            pending_jobs = pending_jobs.sort_values(by="需求日期").reset_index(drop=True)
            job_to_schedule = pending_jobs.iloc[0]

            best_choice = {
                "finish_time": datetime.max,
                "start_time": None,
                "device": None,
                "maintenance_time": None
            }

            # 寻找该工单的最佳设备
            process_key = (job_to_schedule["工艺路线类型"], job_to_schedule["工序号"])
            possible_devices = self.process_routes.get(process_key, [])

            for device_option in possible_devices:
                device_id = device_option["device"]
                
                # 1. 计算加工时间
                processing_time_sec = device_option["time_per_unit"] * job_to_schedule["需求量"]
                processing_duration = timedelta(seconds=int(processing_time_sec))

                # 2. 计算切换时间
                last_prod = devices_status[device_id]["last_product"]
                current_prod = str(job_to_schedule["产品号"])
                switch_key = (device_id, last_prod, current_prod)
                
                switch_duration = timedelta(seconds=0)
                if last_prod is not None and last_prod != current_prod and switch_key in self.switch_times:
                    switch_duration = self.switch_times[switch_key]
                
                total_duration = processing_duration + switch_duration

                # 3. 确定工单就绪时间
                prev_job_id = f"{job_to_schedule['订单号']}-{job_to_schedule['工序号'] - 1}"
                predecessor_finish_time = completed_jobs_time.get(prev_job_id, start_time)
                
                ready_time = max(devices_status[device_id]["available_time"], predecessor_finish_time)
                
                current_day = ready_time.date()
                
                # 4. 检查物料约束
                material_id = job_to_schedule["瓶颈物料号"]
                if pd.notna(material_id):
                    material_id_str = str(int(material_id))
                    while True:
                        supply_key = (material_id_str, current_day)
                        daily_supply = self.material_supply.get(supply_key, 0)
                        
                        usage_key = (material_id_str, current_day)
                        daily_usage = material_usage.get(usage_key, 0)
                        
                        if job_to_schedule["需求量"] <= daily_supply - daily_usage:
                            break # 物料充足
                        else:
                            # 推迟到第二天
                            current_day += timedelta(days=1)
                            ready_time = datetime.combine(current_day, datetime.min.time()) + timedelta(hours=8)
                
                # 5. 寻找可用班次并考虑保养
                # 遍历该设备从 ready_time 开始的班次
                found_slot = False
                for start_shift, end_shift in self.calendar.get(device_id, []):
                    if start_shift.date() < ready_time.date():
                        continue # 跳过已经过去的班次
                    
                    potential_start = max(ready_time, start_shift)
                    
                    # 检查是否需要插入保养
                    maintenance_duration = timedelta(minutes=30)
                    maintenance_info = None
                    if (device_id, potential_start.date()) not in maintenance_scheduled:
                         # 必须在上午12点前保养
                        maintenance_start = max(start_shift, datetime.combine(potential_start.date(), datetime.min.time()) + timedelta(hours=8))
                        maintenance_end = maintenance_start + maintenance_duration
                        if maintenance_end <= (datetime.combine(potential_start.date(), datetime.min.time()) + timedelta(hours=12)) and (maintenance_end <= end_shift):
                            potential_start = max(potential_start, maintenance_end)
                            maintenance_info = (maintenance_start, maintenance_end)
                        else:
                             # 如果无法在上午保养，则该班次不可用
                            continue

                    # 检查班次剩余时间是否足够
                    if potential_start + total_duration <= end_shift:
                        # 找到了合适的班次
                        if potential_start + total_duration < best_choice["finish_time"]:
                            best_choice["finish_time"] = potential_start + total_duration
                            best_choice["start_time"] = potential_start
                            best_choice["device"] = device_id
                            best_choice["maintenance_time"] = maintenance_info
                        found_slot = True
                        break # 已为此设备找到最早的班次，跳出班次循环
                
                if found_slot:
                    continue # 继续检查下一个可选设备

            # 6. 更新状态并记录结果
            if best_choice["device"] is not None:
                # 如果有保养任务，先记录
                if best_choice["maintenance_time"]:
                    m_start, m_end = best_choice["maintenance_time"]
                    maintenance_job = {
                        "订单号": best_choice["device"],
                        "工单号": f"{best_choice['device']}-1",
                        "工序号": 1,
                        "设备号": best_choice["device"],
                        "产品号": "8888",
                        "开始时间": m_start.strftime("%Y-%m-%d %H:%M:%S"),
                        "结束时间": m_end.strftime("%Y-%m-%d %H:%M:%S")
                    }
                    schedule_result.append(maintenance_job)
                    maintenance_scheduled.add((best_choice["device"], m_start.date()))
                
                # 记录工单排产结果
                final_job = {
                    "订单号": job_to_schedule["订单号"],
                    "工单号": job_to_schedule["工单号"],
                    "工序号": job_to_schedule["工序号"],
                    "设备号": best_choice["device"],
                    "产品号": job_to_schedule["产品号"],
                    "开始时间": best_choice["start_time"].strftime("%Y-%m-%d %H:%M:%S"),
                    "结束时间": best_choice["finish_time"].strftime("%Y-%m-%d %H:%M:%S")
                }
                schedule_result.append(final_job)
                
                # 更新设备状态
                devices_status[best_choice["device"]]["available_time"] = best_choice["finish_time"]
                devices_status[best_choice["device"]]["last_product"] = str(job_to_schedule["产品号"])
                
                # 更新物料消耗
                material_id = job_to_schedule["瓶颈物料号"]
                if pd.notna(material_id):
                    material_id_str = str(int(material_id))
                    usage_key = (material_id_str, best_choice["start_time"].date())
                    material_usage[usage_key] = material_usage.get(usage_key, 0) + job_to_schedule["需求量"]
                
                # 更新已完成工单
                completed_jobs_time[job_to_schedule["工单号"]] = best_choice["finish_time"]
                
                # 将下一道工序加入待排产池
                next_seq = job_to_schedule["工序号"] + 1
                next_job = self.data["产品需求"][
                    (self.data["产品需求"]["订单号"] == job_to_schedule["订单号"]) &
                    (self.data["产品需求"]["工序号"] == next_seq)
                ]
                if not next_job.empty:
                    pending_jobs = pd.concat([pending_jobs, next_job], ignore_index=True)
            
            # 从待排产池中移除已处理的工单
            pending_jobs = pending_jobs.drop(0).reset_index(drop=True)

        return pd.DataFrame(schedule_result)


# --- 主程序入口 ---
if __name__ == '__main__':
    # 加载数据
    try:
        data_files = {
            "产品需求": pd.read_csv("初赛/产品需求.csv"),
            "切换时间": pd.read_csv("初赛/切换时间.csv"),
            "工作日历": pd.read_csv("初赛/工作日历.csv"),
            "工艺路线": pd.read_csv("初赛/工艺路线.csv"),
            "瓶颈物料": pd.read_csv("初赛/瓶颈物料.csv")
        }
        print("数据加载成功。")
    except FileNotFoundError as e:
        print(f"错误：找不到数据文件 {e.filename}。请确保文件路径正确。")
        exit()

    # 创建调度器实例并运行
    scheduler = ProductionScheduler(data_files)
    print("开始进行排产调度...")
    final_schedule = scheduler.schedule(start_time_str="2025-05-06 08:00:00")
    
    # 保存结果
    output_filename = "submission_result.csv"
    final_schedule.to_csv(output_filename, index=False, encoding="utf8")
    
    print(f"排产完成，结果已保存到 {output_filename}")
    print("\n排产结果预览:")
    print(final_schedule.head())
```
